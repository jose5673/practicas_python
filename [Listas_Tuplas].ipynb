## ğŸ” **Â¿QuÃ© son las Listas?**

Las **listas** son estructuras de datos **mutables** que pueden almacenar mÃºltiples elementos de cualquier tipo. Son como "contenedores dinÃ¡micos" que pueden crecer, reducirse y modificarse durante la ejecuciÃ³n del programa.

### ğŸ­ **AnalogÃ­a Industrial:**
Imagina una **lÃ­nea de producciÃ³n** donde:
- Puedes **agregar** nuevos productos
- Puedes **quitar** productos defectuosos  
- Puedes **modificar** productos existentes
- Puedes **inspeccionar** cualquier producto por su posiciÃ³n

### âš¡ **CaracterÃ­sticas Fundamentales:**

| CaracterÃ­stica | DescripciÃ³n | Ejemplo Industrial |
|---------------|-------------|-------------------|
| **Mutables** | Se pueden modificar | Agregar nuevo sensor a la red |
| **Ordenadas** | Mantienen secuencia de inserciÃ³n | Orden de lecturas temporales |
| **Indexadas** | Acceso por posiciÃ³n (0, 1, 2...) | Sensor[0] = primer sensor |
| **Duplicados** | Permiten elementos repetidos | MÃºltiples lecturas iguales |
| **HeterogÃ©neas** | Diferentes tipos de datos | [ID, valor, timestamp, estado] |
