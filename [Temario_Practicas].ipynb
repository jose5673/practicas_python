````python
# ğŸ­ EJERCICIO 1.1.2: Conversiones de Tipos en Contexto Industrial
print("\nğŸ¯ EJERCICIO 1.1.2: Conversiones de Tipos")
print("=" * 45)

# âœï¸ ESCENARIO: Recibes datos de sensores como strings desde un sistema SCADA
# Necesitas convertirlos a los tipos correctos para procesamiento

datos_scada = "25.7"    # Temperatura como string
contador_str = "1500"   # Contador como string
estado_str = "1"        # Estado como string (1=ON, 0=OFF)

# ğŸ”§ TU TAREA: Convierte estos datos a los tipos apropiados
temperatura = float(datos_scada)        # string â†’ float
contador = int(contador_str)           # string â†’ int
estado_bool = bool(int(estado_str))    # string â†’ int â†’ bool

print(f"ğŸ“Š Datos originales (strings): {datos_scada}, {contador_str}, {estado_str}")
print(f"ğŸ”„ Datos convertidos:")
print(f"   ğŸŒ¡ï¸  Temperatura: {temperatura} (tipo: {type(temperatura).__name__})")
print(f"   ğŸ“ˆ Contador: {contador} (tipo: {type(contador).__name__})")
print(f"   âš¡ Estado: {estado_bool} (tipo: {type(estado_bool).__name__})")

# âœ… BONUS: ConversiÃ³n con validaciÃ³n
try:
    temp_validada = float(datos_scada)
    if 0 <= temp_validada <= 100:
        print(f"âœ… Temperatura vÃ¡lida: {temp_validada}Â°C")
    else:
        print("âš ï¸ Temperatura fuera de rango")
except ValueError:
    print("âŒ Error: No se pudo convertir la temperatura")
````

### ğŸ¥ˆ **NIVEL 2: INTERMEDIO - Operadores y Expresiones** *(30-60 minutos)*
**Objetivo:** Combinar variables y operadores para cÃ¡lculos industriales complejos

````python
# ğŸ­ EJERCICIO 1.2.1: Calculadora de Eficiencia EnergÃ©tica
print("ğŸ¯ EJERCICIO 1.2.1: Calculadora de Eficiencia EnergÃ©tica")
print("=" * 60)

# âœï¸ ESCENARIO: Calcular la eficiencia de un motor industrial
# FÃ³rmula: Eficiencia (%) = (Potencia de Salida / Potencia de Entrada) Ã— 100

potencia_entrada = 1500  # Watts
potencia_salida = 1275   # Watts

# ğŸ”§ TU TAREA: Calcula la eficiencia y clasifÃ­cala
eficiencia = (potencia_salida / potencia_entrada) * 100

print(f"âš¡ Potencia de Entrada: {potencia_entrada} W")
print(f"ğŸ”Œ Potencia de Salida: {potencia_salida} W")
print(f"ğŸ“Š Eficiencia: {eficiencia:.2f}%")

# ClasificaciÃ³n de eficiencia
if eficiencia >= 95:
    clasificacion = "A+ (Excelente)"
    color = "ğŸŸ¢"
elif eficiencia >= 90:
    clasificacion = "A (Muy Buena)"
    color = "ğŸŸ¢"
elif eficiencia >= 85:
    clasificacion = "B (Buena)"
    color = "ğŸŸ¡"
elif eficiencia >= 80:
    clasificacion = "C (Regular)"
    color = "ğŸŸ "
else:
    clasificacion = "D (Deficiente)"
    color = "ğŸ”´"

print(f"{color} ClasificaciÃ³n: {clasificacion}")

# âœ… CÃLCULOS ADICIONALES
perdida_energia = potencia_entrada - potencia_salida
costo_perdida_diaria = perdida_energia * 24 * 0.15  # $0.15 por kWh

print(f"\nğŸ“‰ PÃ©rdida de EnergÃ­a: {perdida_energia} W")
print(f"ğŸ’° Costo de PÃ©rdida Diaria: ${costo_perdida_diaria/1000:.2f}")

# ğŸ¯ VALIDACIONES DE SEGURIDAD
temperatura_critica = 85  # Â°C
temperatura_actual = 78   # Â°C

sobrecalentamiento = temperatura_actual > temperatura_critica
if sobrecalentamiento:
    print("ğŸš¨ ALERTA: Motor sobrecalentado!")
else:
    print("âœ… Temperatura del motor normal")
````
